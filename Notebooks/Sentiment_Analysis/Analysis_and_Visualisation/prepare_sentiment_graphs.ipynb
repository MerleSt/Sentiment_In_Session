```python
# change in sentiment over time
# average sentiment for each date?
# groupbydate and mean
average_sent_over_time_pos = df.groupby(df.Date.dt.month)['positive'].mean()
average_sent_over_time_neg = df.groupby(df.Date.dt.month)['negative'].mean()
average_sent_over_time_neu = df.groupby(df.Date.dt.month)['neutral'].mean()
```

```python
# Create dataframes
df_positive = pd.DataFrame({'Date': average_sent_over_time_pos.index.to_list(), 'Average Sentiment (Positive)': average_sent_over_time_pos.values})
df_negative = pd.DataFrame({'Date': average_sent_over_time_neg.index.to_list(), 'Average Sentiment (Negative)': average_sent_over_time_neg.values})
df_neutral = pd.DataFrame({'Date': average_sent_over_time_neu.index.to_list(), 'Average Sentiment (Neutral)': average_sent_over_time_neu.values})
```

```python
# Merge dataframes on 'Date' column
merged_df = pd.merge(df_positive, df_negative, on='Date', how='outer')
merged_df = pd.merge(merged_df, df_neutral, on='Date', how='outer')
```
